**Konversi Tipe Data**

Kenapa data butuh dikonversi?
Karena tidak semua data valid dan dapat diolah langsung sesuai kebutuhan.

```python
#Python mendeteksi usia sebagai str, bukan int
usia = input('Masukkan usia anda: ')
print(type(usia))
```

```python
#Konversi Tipe Data Secara Implisit
a = 10/2
print(type(a))
b = 1/2
print(type(b))

#ex.
kata = 'jeruk '*5
print(kata)
```

```python
#Konversi Tipe Data Secara Eksplisit
#int()
#float()
#str()
#bool()

#----Konversi String Ke Numerik----#
panjang_str = input('Masukkan panjang: ')
lebar_str = input('Masukkan lebar: ')

# Convert to numeric types before multiplication
panjang = int(panjang_str)
lebar = float(lebar_str)
print('Luas = ', panjang * lebar)
```

```python
panjang = int(input('Masukkan panjang:'))
lebar = float(input('Masukkan lebar: '))
print('Luas = ', panjang * lebar)
```

```python
#----Konversi Numerik Ke String----#
nama = 'wudi'
tahun_lahir = 2000
print(nama + ' lahir di tahun ' + str(tahun_lahir))
```
